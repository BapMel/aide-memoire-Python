## Fonctions

Une fonction est une suite d'instructions qui dépend de paramètres.

```python
def hypoténuse(x,y):
    z = math.sqrt(x*x + y*y)
    return z
```

```python
hypoténuse(3.0,4.0)
```

```python
def puissance(x,n):
    """Calcule x à la puissance n, en supposant x > 0 et n >= 0"""
    if x == 0:
        return 0
    r = 1
    for i in range(n):
        r *= x
    return r
```

```python
puissance(2,10)
```

Fonctions locales :

```python
def max3(x,y,z):
    def max2(u,v):
        if u > v:
            return u
        else:
            return v
    return max2(x, max2(y,z))
```

Fonctions comme valeurs de première classe :

```python
def somme_fonction(f,n):
    s = 0
    for i in range(n+1):
        s = s + f(i)
    return s

def carré(x):
    return x*x

somme_fonction(carré, 10)
```

Fonction anonyme :

```python
somme_fonction(lambda x: x*x, 10)
```

Renvoyer une fonction comme résultat :

```python
def composee(f,g):
    def h(x):
        return f(g(x))
    return h

def composee_autre_définition(f,g):
    return lambda x: f(g(x))

a = composee(carré,carré)
a(4)
```

Fonction dérivée :

```python
def dérive(f,epsilon):
    return lambda x: (f(x+epsilon) -f(x))/epsilon
```

```python
a = dérive(carré,0.001)
```

```python
a(2)
```

```python
a(5)
```

## Récursivité

```python
def factorielle(n):
    assert x >= 0
    if n == 0:
        return 1
    else:
        return n * factorielle(n-1)
```

```python
factorielle(4)
```

```python
factorielle(10)
```

```python
def puissance_rec(x,n):
    assert n >= 0
    if n == 0:
        return 1
    else:
        return x * puissance_rec(x, n-1)
```

```python
puissance_rec(2,10)
```

Recherche dans un tableau :

```python
def indice(x,s):
    for i, y in enumerate(s):
        if y == x:
            return i
    return None
```

```python
indice(1, [1,2,3])
```
